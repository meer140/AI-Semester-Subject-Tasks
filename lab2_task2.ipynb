```python
class Trapezoid:
    def __init__(self, a, b, h):
        self.a = a  
        self.b = b  
        self.h = h

    def area(self):
        return 0.5 * (self.a + self.b) * self.h


class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return self.base * self.height


shapes_data = [
    ("Trapezoid", Trapezoid(10, 6, 4)),
    ("Parallelogram", Parallelogram(8, 5)),
    ("Trapezoid", Trapezoid(12, 8, 6)),
    ("Parallelogram", Parallelogram(10, 7))
]


areas = {}

for i, (name, obj) in enumerate(shapes_data, start=1):
    areas[f"{name}_{i}"] = obj.area()


print("All Shape Areas:")
for k, v in areas.items():
    print(k, ":", v)


max_shape = max(areas, key=areas.get)

print("\nShape with Maximum Area:")
print(max_shape, "=", areas[max_shape])
```
