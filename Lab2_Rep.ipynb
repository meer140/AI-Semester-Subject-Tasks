```python
marks = int(input("Enter marks (1-100): "))

if marks < 1 or marks > 100:
    print("Invalid input")
elif marks < 50:
    print("Grade: F")
elif marks <= 60:
    print("Grade: E")
elif marks <= 70:
    print("Grade: D")
elif marks <= 80:
    print("Grade: C")
elif marks <= 90:
    print("Grade: B")
else:
    print("Grade: A")
```

```python
n = int(input("Enter number of terms: "))

if n <= 0:
    print("Invalid input")
else:
    a = 0
    b = 1
    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# --- Step 1: Create data ---
categories = np.array(['A', 'B', 'C', 'D', 'E'])      # Category labels
values = np.array([10, 24, 36, 18, 30])               # Values for each category

# --- Step 2: Vertical Bar Graph ---
plt.figure(figsize=(8,5))
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Graph')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Step 3: Horizontal Bar Graph ---
plt.figure(figsize=(8,5))
plt.barh(categories, values, color='orange', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Graph')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
numbers = [5, 2, 9, 1, 7]

print(len(numbers))      # Length of list: 5
print(min(numbers))      # Smallest: 1
print(max(numbers))      # Largest: 9
print(sum(numbers))      # Sum: 24
numbers.sort()           # Sort ascending
print(numbers)           # [1, 2, 5, 7, 9]
numbers.reverse()        # Reverse order
print(numbers)           # [9, 7, 5, 2, 1]
```

```python
exam1_marks = {85, 90, 78, 92, 88,90}
exam2_marks = {80, 90, 88, 95, 75}

# Print sets
print("Exam 1 marks:", exam1_marks)
print("Exam 2 marks:", exam2_marks)

# Add a new mark
exam1_marks.add(100)
print("Exam 1 after adding 100:", exam1_marks)

# Remove a mark
exam2_marks.remove(75)
print("Exam 2 after removing 75:", exam2_marks)
all_marks = exam1_marks.union(exam2_marks)
print("All unique marks:", all_marks)


common_marks = exam1_marks.intersection(exam2_marks)
print("Marks in both exams:", common_marks)

unique_exam1 = exam1_marks.difference(exam2_marks)
print("Marks only in Exam 1:", unique_exam1)


print("Is 92 in Exam 1?", 92 in exam1_marks)
print("Is 75 in Exam 2?", 75 in exam2_marks)
```
