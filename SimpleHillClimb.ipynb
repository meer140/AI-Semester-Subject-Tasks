```python
# -------------------------------------------------
# Hill Climbing Algorithm (Detailed Version)
# -------------------------------------------------

import random

# -------------------------------------------------
# Objective Function
# Peak value occurs near x = 15
# -------------------------------------------------
def objective_function(x):
    return -(x - 15) ** 2 + 225


# -------------------------------------------------
# Generate Neighboring States
# -------------------------------------------------
def get_neighbors(current_state):
    neighbors = []

    # left neighbor
    neighbors.append(current_state - 1)

    # right neighbor
    neighbors.append(current_state + 1)

    return neighbors


# -------------------------------------------------
# Hill Climbing Algorithm
# -------------------------------------------------
def hill_climbing(start_state):

    current_state = start_state
    current_value = objective_function(current_state)

    step = 1

    print("======================================")
    print("     HILL CLIMBING ALGORITHM")
    print("======================================")

    print("\nStarting State :", current_state)
    print("Starting Value :", current_value)

    print("\nSearching for better states...\n")

    while True:

        print("--------------------------------------")
        print("STEP :", step)
        print("--------------------------------------")

        # Generate neighbors
        neighbors = get_neighbors(current_state)

        print("Current State :", current_state)
        print("Current Value :", current_value)

        print("\nGenerated Neighbors:")

        for n in neighbors:
            print("Neighbor =", n,
                  " Objective Value =", objective_function(n))

        # Find best neighbor
        best_neighbor = neighbors[0]
        best_value = objective_function(best_neighbor)

        for neighbor in neighbors:

            value = objective_function(neighbor)

            if value > best_value:
                best_neighbor = neighbor
                best_value = value

        print("\nBest Neighbor Found :", best_neighbor)
        print("Best Neighbor Value :", best_value)

        # Check if improvement exists
        if best_value <= current_value:

            print("\nNo better neighbor found.")
            print("Algorithm stops here.")
            print("\nReached Peak / Local Maximum")

            break

        # Move to better state
        print("\nMoving to better state...")

        current_state = best_neighbor
        current_value = best_value

        step += 1

    print("\n======================================")
    print(" FINAL RESULT ")
    print("======================================")

    print("Best State Found :", current_state)
    print("Best Objective Value :", current_value)

    return current_state


# -------------------------------------------------
# Main Program
# -------------------------------------------------
print("Program Started...\n")

# Random starting point
start = random.randint(0, 30)

# Run Hill Climbing
hill_climbing(start)

print("\nProgram Finished.")
```
