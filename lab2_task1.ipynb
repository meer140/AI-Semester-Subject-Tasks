```python
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
#Create object
s1 = Student("mir",20)

#display namespace(attributes of instance)

print(s1.__dict__)
```

```python
def student_data(student_id, student_name=None, student_class=None):
    print("Student ID:", student_id)

    if student_name:
        print("Student Name:", student_name)

    if student_class:
        print("Class:", student_class)


student_data(62)
student_data(62,'Mir')
student_data(62,'mir',12)

```

```python
class Student:
    def __init__(self, sid, name, sclass):
        self.sid = sid
        self.name = name
        self.sclass = sclass

    def display(self):
        for key,value in self.__dict__.items():
            print(key," : ",value)

s1 = Student(62,"mir",12)
s1.display()
```

```python
import sys

print(f"Size of object {sys.getsizeof(s1)}")
print(f"Size of attributes {sys.getsizeof(s1.__dict__)}")
```

```python
class Student:

    __slots__ = ['name', 'marks','sclass']

    def __init__(self, name, marks,sclass):
        self.name = name
        self.marks = marks
        self.sclass = sclass

s = Student("Ali", 85,12)

print("Memory with slots:", sys.getsizeof(s))
```
